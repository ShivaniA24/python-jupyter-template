### Exercise 3: Scatter Plot with Annotations

```python
# TASK: Create a scatter plot with x = [1, 2, 3, 4, 5] and y = [2, 4, 6, 8, 10].
# Annotate each point with its (x, y) value, and set the title as 'Scatter Plot with Annotations'.
```

```python
import matplotlib.pyplot as plt

# Data
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

# Create the scatter plot
plt.figure(figsize=(8, 6))
plt.scatter(x, y, color='purple', s=100, label='Data Points')

# Annotate each point
for i, (x_val, y_val) in enumerate(zip(x, y)):
    plt.text(x_val, y_val + 0.3, f'({x_val}, {y_val})', fontsize=10, ha='center')

# Add title and labels
plt.title('Scatter Plot with Annotations', fontsize=16)
plt.xlabel('x values', fontsize=14)
plt.ylabel('y values', fontsize=14)
plt.legend(fontsize=12)
plt.grid(True, linestyle='--', alpha=0.7)

# Show the plot
plt.show()
```
