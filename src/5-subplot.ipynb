### Exercise 5: Subplot Layout

```python
# TASK: Create a 2x2 subplot layout.
# Plot a line chart in the first subplot, a bar chart in the second, a scatter plot in the third, and a pie chart in the fourth.
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Data for the plots
x = np.linspace(0, 10, 100)  # Line chart data
y = x**2

categories = ['A', 'B', 'C', 'D']  # Bar chart data
values = [5, 7, 3, 9]

scatter_x = [1, 2, 3, 4, 5]  # Scatter plot data
scatter_y = [2, 4, 6, 8, 10]

pie_labels = ['Python', 'Java', 'C++', 'JavaScript']  # Pie chart data
pie_sizes = [40, 25, 20, 15]

# Create the 2x2 subplot layout
fig, axes = plt.subplots(2, 2, figsize=(10, 8))

# Line plot (1st subplot)
axes[0, 0].plot(x, y, label='y = x^2', color='blue', linewidth=2)
axes[0, 0].set_title('Line Plot')
axes[0, 0].set_xlabel('x values')
axes[0, 0].set_ylabel('y values')
axes[0, 0].grid(True)
axes[0, 0].legend()

# Bar plot (2nd subplot)
axes[0, 1].bar(categories, values, color=['red', 'green', 'blue', 'orange'])
axes[0, 1].set_title('Bar Plot')
axes[0, 1].set_xlabel('Categories')
axes[0, 1].set_ylabel('Values')

# Scatter plot (3rd subplot)
axes[1, 0].scatter(scatter_x, scatter_y, color='purple', s=100)
for i, (x_val, y_val) in enumerate(zip(scatter_x, scatter_y)):
    axes[1, 0].text(x_val, y_val + 0.3, f'({x_val}, {y_val})', fontsize=9, ha='center')
axes[1, 0].set_title('Scatter Plot')
axes[1, 0].set_xlabel('x values')
axes[1, 0].set_ylabel('y values')

# Pie chart (4th subplot)
axes[1, 1].pie(pie_sizes, labels=pie_labels, autopct='%1.1f%%', startangle=140, 
               colors=['gold', 'skyblue', 'lightgreen', 'coral'])
axes[1, 1].set_title('Pie Chart')

# Adjust layout
plt.tight_layout()

# Show the plots
plt.show()
```
